```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
X, y = make_classification(n_samples=100, n_features=2, n_informative=1, n_redundant=0,
            n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=20)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
plt.show()
```

```python
def step(num):
    return 1 if num > 0 else 0
def perceptron(x, y_):
    # Make a new column at start for biases and insert 1 in the whole column.
    x = np.insert(x, 0, 1, axis=1)
    # Initialize weights array which has same size as the input columns.
    weights = np.ones(x.shape[1])
    # Learning rate
    lr = 0.1

    for i in range(1000):
        # Selecting a random row from the input data.
        j = np.random.randint(0, 100)
        # Predict on the basis of selected row.
        y_hat = step(np.dot(x[j], weights))
        # Apply formula.
        weights = weights + lr * (y_[j] - y_hat) * x[j]
    # return intercept and coefficients.
    return weights[0], weights[1:]
```

```python
intercept_, coef_ = perceptron(X, y)
```

```python
print(intercept_)
print(coef_)
```

```python
m = -(coef_[0] / coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3, 3, 100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
plt.ylim(-3, 2)
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(X, y)
```

```python
m = -(lor.coef_[0][0] / lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input_1 = np.linspace(-3, 3, 100)
y_input_1 = m*x_input + b
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.plot(x_input_1, y_input_1, color='black', linewidth=3)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter', s=100)
plt.ylim(-3, 2)
plt.show()
```
